## **PCA**

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Separate features and target
X = df.drop(columns=['is_fraud'])
y = df['is_fraud']

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a PCA instance and fit it to the scaled data
pca = PCA()  # You can specify the number of components here if needed
pca.fit(X_scaled) # Fit the PCA model to your data


# Get the names of the original columns
feature_names = X.columns

# Explained variance ratio for each feature
explained_variance_ratio = pca.explained_variance_ratio_


# Plot feature importance with column names
plt.figure(figsize=(12, 6))
sns.barplot(x=feature_names, y=explained_variance_ratio, hue=feature_names, palette="viridis", legend=False)
# plt.title("Feature Importance by Explained Variance Ratio")
plt.xlabel("Features")
plt.ylabel("Explained Variance Ratio")
plt.xticks(rotation=45, ha="right", fontweight='bold')  # Make x-axis tick labels bold
plt.yticks(fontweight='bold')  # Make y-axis tick labels bold
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
# Step 3: 3D Scatter Plot using the top contributing features
# Get the top contributing features
top_features_indices = np.argsort(-explained_variance_ratio)[:2]  # Indices of the top 2 features
top_features = X.columns[top_features_indices]  # Names of the top 2 features

# Combine original features with the target
df_top_features = pd.DataFrame(X_scaled[:, top_features_indices], columns=top_features)
df_top_features['is_fraud'] = y.values

# Plot the 3D scatter
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
scatter = ax.scatter(
    df_top_features[top_features[0]],
    df_top_features[top_features[1]],
    df_top_features['is_fraud'],
    c=df_top_features['is_fraud'], cmap='coolwarm', alpha=0.7
)

# Labels and title
ax.set_title('3D Scatter Plot of Top Features')
ax.set_xlabel(top_features[0])
ax.set_ylabel(top_features[1])
ax.set_zlabel('Fraud Status')

# Add color bar for fraud status
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Fraud Status')

# Enable rotation
plt.show()
```

# **Machine Learning Model**

```python


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    classification_report, confusion_matrix, ConfusionMatrixDisplay, RocCurveDisplay
)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, LSTM, GRU, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

import sklearn.metrics as metrics
from sklearn.metrics import classification_report, confusion_matrix
# Define a dictionary to store models and their results
```

## **Train Test Split**

```python

X = df.drop(columns=['is_fraud'])
y = df['is_fraud']
```

```python
X
```

```python
X.shape
```

```python

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python

# Normalize continuous features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## **Initialize  the Mahine Learning Model**

```python
import time
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay

# Define a dictionary to store models and their results

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
}

```

## **Evalute the Machine Learning Model**

```python
import time
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay

# Train and evaluate each model
for model_name, model in models.items():
    print(f"Training and evaluating: {model_name}")

    # Measure training time
    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time

    # Measure prediction time
    start_time = time.time()
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    pred_time = time.time() - start_time

    # Calculate evaluation metrics
    accuracy = metrics.accuracy_score(y_test, y_pred)
    precision = metrics.precision_score(y_test, y_pred)
    recall = metrics.recall_score(y_test, y_pred)
    f1_score = metrics.f1_score(y_test, y_pred)
    roc_auc = metrics.roc_auc_score(y_test, y_proba)

    # Print evaluation metrics
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1_score:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    print(f"Training Time: {train_time:.4f} seconds")
    print(f"Prediction Time: {pred_time:.4f} seconds")
    print("-" * 50)

    # Display classification report
    print(f"Classification Report for {model_name}:\n")
    print(classification_report(y_test, y_pred))
    print("-" * 50)

    # Display confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(cmap="Blues", values_format='g') # Format values as integers
    plt.title(f"Confusion Matrix for {model_name}", fontweight='bold') # Bold title
    plt.xlabel(xlabel="Predicted Label", fontweight='bold') # Bold x-axis label
    plt.ylabel(ylabel="True Label", fontweight='bold') # Bold y-axis label

    # Bold values inside the confusion matrix
    for text in disp.text_.ravel():
        text.set_fontweight('bold')
    plt.show()

# Plot ROC curve
    RocCurveDisplay.from_estimator(model, X_test, y_test)
    plt.title(f"ROC Curve for {model_name}", fontweight='bold')  # Make title bold
    plt.xlabel("False Positive Rate (FPR)", fontweight='bold') # Make x-axis label bold
    plt.ylabel("True Positive Rate (TPR)", fontweight='bold') # Make y-axis label bold
    plt.xticks(fontweight='bold') # Make x-axis ticks bold
    plt.yticks(fontweight='bold') # Make y-axis ticks bold
    plt.show()

    print("=" * 70)
```

# **Deep Learning model**

```python

# Function to evaluate the model and display metrics and plots
def evaluate_model(model, X_test, y_test, history=None, train_time=None):
    # Predict on the test set
    start_time = time.time()
    y_pred_proba = model.predict(X_test).flatten()
    y_pred = (y_pred_proba > 0.5).astype(int)
    pred_time = time.time() - start_time

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # Print metrics
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    if train_time is not None:
        print(f"Training Time: {train_time:.4f} seconds")
    print(f"Prediction Time: {pred_time:.4f} seconds")

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:")
    print(cm)

    # Print classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Display confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(cmap="Blues", values_format='g') # Format values as integers
    plt.title(f"Confusion Matrix for {model_name}", fontweight='bold') # Bold title
    plt.xlabel(xlabel="Predicted Label", fontweight='bold') # Bold x-axis label
    plt.ylabel(ylabel="True Label", fontweight='bold') # Bold y-axis label

    # Bold values inside the confusion matrix
    for text in disp.text_.ravel():
        text.set_fontweight('bold')
    plt.show()

    # Plot ROC curve
    RocCurveDisplay.from_estimator(model, X_test, y_test)
    plt.title(f"ROC Curve for {model_name}", fontweight='bold')  # Make title bold
    plt.xlabel("False Positive Rate (FPR)", fontweight='bold') # Make x-axis label bold
    plt.ylabel("True Positive Rate (TPR)", fontweight='bold') # Make y-axis label bold
    plt.xticks(fontweight='bold') # Make x-axis ticks bold
    plt.yticks(fontweight='bold') # Make y-axis ticks bold
    plt.show()

    # Plot training history if provided
    if history is not None:
        plt.figure(figsize=(12, 6))
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title("Training and Validation Loss")
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
        plt.legend()
        plt.show()

```

```python

input_shape = (X_train.shape[1], 1) # Reshape to (num_features, 1) for LSTM
```

## **LSTM Model**

### **Build the LSTM model**

```python


# Function to build the LSTM model
def build_lstm_model(input_shape):
    model = Sequential([
        LSTM(128, activation='relu', input_shape=input_shape, return_sequences=True),
        Dropout(0.3),
        LSTM(64, activation='relu', return_sequences=False),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')  # Output layer for binary classification
    ])
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model


# Build and train LSTM model
lstm_model = build_lstm_model(input_shape)
```

### **Train LSTM model**

```python

start_time = time.time()

history_lstm = lstm_model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
train_time_lstm = time.time() - start_time
```

### **Evaluate LSTM model**

```python

# Evaluate LSTM model
evaluate_model(lstm_model, X_test, y_test, history=history_lstm, train_time=train_time_lstm)
```

## **GRU Model**

### **Build the GRU model**

```python

# Function to build the GRU model
def build_gru_model(input_shape):
    model = Sequential([
        GRU(128, activation='relu', input_shape=input_shape, return_sequences=True),
        Dropout(0.3),
        GRU(64, activation='relu', return_sequences=False),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')  # Output layer for binary classification
    ])
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
```

### **Train GRU model**

```python

# Build and train GRU model
gru_model = build_gru_model(input_shape)
start_time = time.time()
history_gru = gru_model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,

    verbose=1
)
train_time_gru = time.time() - start_time
```


### **Evaluate GRU model**

```python

# Evaluate GRU model
evaluate_model(gru_model, X_test, y_test, history=history_gru, train_time=train_time_gru)
```
